```r
inner_join(reservation_r,order_info_r,by="RESERV_NO") %>% 
arrange(RESERV_NO,ITEM_ID) %>% 
select(RESERV_NO,CUSTOMER_ID,VISITOR_CNT,CANCEL,ORDER_NO,ITEM_ID,SALES)
```

```r
left_join(reservation_r,order_info_r,by="RESERV_NO") %>% 
arrange(RESERV_NO,ITEM_ID) %>% 
select(RESERV_NO,CUSTOMER_ID,VISITOR_CNT,CANCEL,ORDER_NO,ITEM_ID,SALES)
```

```r
table_added_row <- order_info_r %>% 
add_row(ORDER_NO="1",ITEM_ID="1",RESERV_NO="1",SALES=1)
full_join(reservation_r,table_added_row,by="RESERV_NO") %>% 
arrange(RESERV_NO,ITEM_ID) %>% 
select(RESERV_NO,CUSTOMER_ID,VISITOR_CNT,CANCEL,ORDER_NO,ITEM_ID,SALES)
```

```r
reservation_r %>% 
group_by(CUSTOMER_ID) %>% 
summarise(avg=mean(VISITOR_CNT)) %>% 
filter(avg>=3) %>% 
arrange(desc(avg))
```

```r
my_first_cook <- order_info_r %>% 
mutate(reserv_month=substr(RESERV_NO,1,6)) %>% 
group_by(ITEM_ID,reserv_month) %>% 
summarise(avg_sales=mean(SALES)) %>% 
arrange(ITEM_ID,reserv_month)

my_first_cook
```

```r
# 퀴즈
mpg <- as.data.frame(ggplot2::mpg)
head(mpg)
```

```r
# 배기량(displ)이 4이하인 자동차와 5이상인 자동차 중
# 어떤 자동차의 고속도로 연비(hwy)가 평균적으로 더 높은지?

# 평균 연비
#avg_hwt <- mean(hwy))

mpg %>% count()
```

```r
summarise(mpg,avg_hwt=mean(hwy))
```

```r
# displ이 4 이하인 행
displ_4 <- mpg %>% filter(displ<=4)
displ_4
```

```r
# displ이 5 이상인 행
displ_5 <- mpg %>% filter(displ<=4)
displ_5
```

```r
avg_hwt1=mean(displ_4)
avg_hwt2=mean(displ_5)
avg_hwt1
avg_hwt2
```

```r
#해답편
displ_4 <- mpg %>% filter(displ<=4)
displ_5 <- mpg %>% filter(displ>=5)
# 위까지는 맞았음
mean(displ_4$hwy) # 추출한 데이터의 hwy들의 평균
mean(displ_5$hwy)
```

```r
# 퀴즈 2

manuf_audi <- mpg %>% filter(manufacturer=="audi") # 자동차 제조 회사가 audi
manuf_toyoda <- mpg %>% filter(manufacturer=="toyota") # 자동차 제조 회사가 toyoda

mean(manuf_audi$cty)
mean(manuf_toyoda$cty)
```
